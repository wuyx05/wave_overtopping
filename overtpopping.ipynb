```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df=scaler.fit_transform(data[data.columns[~(data.columns.isin( ["Name",'q']))]])
df = pd.DataFrame(df,columns=data.columns[~(data.columns.isin( ["Name",'q']))]).reset_index(drop=True)
```

```python
df = pd.concat([df,pd.to_numeric(data.q.reset_index(drop=True))],axis=1)
```

```python
import lightgbm as lgb
```

```python
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python

df = df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[df.columns[~df.columns.isin(['q'])]], df.q, test_size=0.2, random_state=42)
```

```python
lgb_train = lgb.Dataset(x_train, y_train,categorical_feature=['RF','CF'])
lgb_eval = lgb.Dataset(x_test, y_test, reference=lgb_train)
```

```python
df.q.dtypes
```

```python
params = {
    'task': 'train', 
    'boosting': 'gbdt',
    'objective': 'regression',
    'num_leaves': 10,
    'learnnig_rage': 0.05,
    'metric': {'l2','l1'},
    'verbose': -1
}
```

```python
model = lgb.train(params,
                 train_set=lgb_train,
                 valid_sets=lgb_eval,
                 early_stopping_rounds=30)
```

```python
# prediction
y_pred = model.predict(x_test)

# accuracy check
mse = mean_squared_error(y_test, y_pred)
rmse = mse**(0.5)
print("MSE: %.2f" % mse)
print("RMSE: %.2f" % rmse) 
```
